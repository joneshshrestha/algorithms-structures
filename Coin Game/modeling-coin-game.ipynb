```python
# Here is a simple game: you have four coins lined up on a table; some
# coins are facing (H)eads up, and some are facing (T)ails up. E.g. TTHT corresponds to four coins,
# of which the first two and the last are tails up, and the third one is heads up. (Better for encoding:
# in binary with 0 = ‘H’ and 1 = ‘T’, we are looking at 1101, that is, the number 13 in decimal.) For
# four coins there are 16 possible configurations, from HHHH (0000 = 0) to TTTT (1111 = 15). In
# the game you are allowed to flip any two coins which are not next to each other. For example,
# starting with TTHT we could flip the first and third coin to get HTTT, or the first and fourth coin,
# to get HTHH. Using networkx to draw a graph, in which the vertices are the states, and edges are the
# legal moves in the game.

import matplotlib.pyplot as plt
import networkx as nx
from itertools import chain, combinations

def coinFlip3():
    'Coin Flip'

    G = nx.Graph()

    for vx in range(10):
        G.add_node(vx)
    
    labels = {0: '0000', 1: '0001', 2: '0010',
              3: '0011', 4: '0100', 5: '0101',
              6: '0110', 7: '0111', 8: '1000',
              9: '1001', 10:'1010', 11: '1011', 
              12: '1100', 13:'1101', 14: '1110', 
              15: '1111'}
    
    G.add_edges_from([
        (0,5), (0,9), (0,10),
        (1,4), (1,8), (1,11),
        (2,7), (2,8), (2,11),
        (3,6), (3,9), (3,10),
        (4,1), (4,13), (4,14),
        (5,0), (5,15), (5,12),
        (6,3), (6,15), (6,12),
        (7,2), (7,14), (7,13),
        (8,2), (8,1), (8,13),
        (9,3), (9,0), (9,12),
        (10,3), (10,0), (10,15),
        (11,2), (11,1), (11,14),
        (12,9), (12,6), (12,5),
        (13,4), (13,7), (13,8),
        (14,4), (14,7), (14,11),
        (15,5), (15,6), (15,10)])
    
    pos = nx.circular_layout(G)
    
    nx.draw_networkx_nodes(G,pos,node_color='orange',node_size = 20)
    nx.draw_networkx_edges(G,pos)
    nx.draw_networkx_labels(G,pos,labels,font_size = 10)

    plt.show() # to display graph
```

```python
coinFlip3()
```

```python
def coin_label (n, k) :
    'return length k H/T representation of int n'
    coins = ''
    for i in range (k):
        coins = ('H' if n % 2 == 0 else 'T') + coins
        n //= 2
    print(coins)
    return coins
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def coinFlip():
    'Coin Flip'

    G = nx.Graph()

    for vx in range(16):
        G.add_node(vx)
    
    labels = {}
    for i in range(16):
        labels[i] = coin_label(i,4)
    
    G.add_edges_from([
        (0,5), (0,9), (0,10),
        (1,4), (1,8), (1,11),
        (2,7), (2,8), (2,11),
        (3,6), (3,9), (3,10),
        (4,1), (4,13), (4,14),
        (5,0), (5,15), (5,12),
        (6,3), (6,15), (6,12),
        (7,2), (7,14), (7,13),
        (8,2), (8,1), (8,13),
        (9,3), (9,0), (9,12),
        (10,3), (10,0), (10,15),
        (11,2), (11,1), (11,14),
        (12,9), (12,6), (12,5),
        (13,4), (13,7), (13,8),
        (14,4), (14,7), (14,11),
        (15,5), (15,6), (15,10)])
    
    pos = nx.circular_layout(G)
    
    nx.draw_networkx_nodes(G,pos,node_color='orange',node_size = 20)
    nx.draw_networkx_edges(G,pos)
    nx.draw_networkx_labels(G,pos,labels,font_size = 10)

    # Highlight specific edges (e.g., between nodes 0 and 1)
    highlighted_edges = [(15, 5), (5, 12), (12, 9)]

    # Draw highlighted edges with different color and width
    nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='red', width=2)

    plt.show() # to display graph
```

```python
from itertools import permutations
def pancake_sort(pancakes):
    
    def flip(arr, k):
        return arr[:k][::-1] + arr[k:]
    
    def find_max_index(arr, k):
        return arr[:k].index(max(arr[:k]))
    
    n = len(pancakes)
    flips = []
    result = pancakes.copy()
    
    for i in range(n, 1, -1):
        # Find largest pancake in the unsorted array portion
        max_index = find_max_index(result, i)
        
        # If largest pancake isn't already in position
        if max_index != i-1:
            # If largest isn't at top, flip to bring it to top
            if max_index != 0:
                result = flip(result, max_index + 1)
                flips.append(max_index + 1)
            
            # Flip to put it in correct position
            result = flip(result, i)
            flips.append(i)
    
    return result, flips

# Example usage and testing
def test_pancake_sort():
    result = []
    for i in lst:  # Fix each element as the first one
        for perm in permutations([x for x in lst if x != i]):  # Permute the rest
            result.append([i] + list(perm))
    test_cases = result
    
    for case in test_cases:
        sorted_pancakes, flips = pancake_sort(case.copy())
        print(f"\nInput:  {case}")
        print(f"Output: {sorted_pancakes}")
        print(f"Number of flips: {len(flips)}")
        print(f"Flip sequence: {flips}")
        # Verify result is sorted
        assert sorted_pancakes == sorted(case), "Sort failed!"

# Run tests
if __name__ == "__main__":
    test_pancake_sort()
```

```python
def flip(lst):

    # helper function that returns the index of the maximum value of an array from 0th position to kth position
    def find_max_index(arr, k):
        return arr[:k].index(max(arr[:k]))

    # initializing commands array 
    commands = []

    # recursive function that takes an array, number of items 
    def recursive_sort(arr, n):
        # base case
        # if only 1 items is left to sort
        if n <= 1:
            return commands
            
        # max value in the unsorted array portion
        max_index = find_max_index(arr, n)
        
        # checks if the max value is not already on the bottom(last) of the array
        if max_index != n-1:
            # if the max value is not in the correct position(last) then checks 
            # if max element is not on the top(first) position either
            if max_index != 0:
                # flip/reverse the array from top(first) position to max value position
                arr[:max_index + 1] = arr[:max_index + 1][::-1]
                # append number of top pancakes that are flipped i.e index 0 to max value index
                commands.append(f"Flip top {max_index + 1} pancakes")
            
            # flipping the unsorted array from index 0 to n, so that max value stays at the bottom of the array
            arr[:n] = arr[:n][::-1]
            # append number of top pancakes that are flipped i.e index 0 to n
            commands.append(f"Flip top {n} pancakes")
        
        # recursion reducing the number of array by 1, since the last item is always sorted
        # the max value is at the bottom of the array i.e last position
        return recursive_sort(arr, n-1)

    # running the recursive sorting function 
    recursive_sort(lst, len(lst))
     
    # print each command from the commands list
    for command in commands:
        print(command)
```

```python
import sys
from random import *

def disp_game(board,n, start, target):
    'display board of size nxn with start and target'
    
    print('+'+(2*n-1)*'-'+'+')
    for y in range(n):
        print('|',end = '')
        for x in range(n):
            if board[(x,y)] == 'N':
                arrow = '\u2191'
            elif board[(x,y)] == 'S':
                arrow = '\u2193'
            elif board[(x,y)] == 'W':
                arrow = '\u2190'
            elif board[(x,y)] == 'E':
                arrow = '\u2192'
            if (x,y) == target:
                sys.stdout.shell.write('X', 'COMMENT')
                print('|', end = '')
            elif (x,y) == start:
                sys.stdout.shell.write(arrow, 'COMMENT')
                print('|', end = '')                
            else:
                print(arrow, end = '|')
        print()
        if y < n-1:
            print('|'+(2*n-1)*'-'+'|')
    print('+'+(2*n-1)*'-'+'+')
    
def create_game(n,k):
    'create random nxn board game which can be solved in at most k steps'
    
    sol = [(0,0)]
    board  = {}
    flip = (random() < 0.5)
    for _ in range(k):
        (x,y) = sol[-1]
        xnew,ynew = x,y
        while (xnew,ynew) == (x,y) or (xnew,ynew) in board:
            if flip:
                xnew = randrange(n)
                board[(x,y)] = 'W' if xnew < x else 'E'
            else:
                ynew = randrange(n)
                board[(x,y)] = 'N' if ynew < y else 'S'
        flip = not flip
        sol.append((xnew,ynew))
    target = (xnew,ynew)
    for x in range(n):
        for y in range(n):
            if (x,y) not in board:
                board[(x,y)] = choice('NSWE')
    print('Move from (0,0) to {}'.format(target))
    disp_game(board,n,(0,0),target)
    print('There is a solution in {} steps'.format(k))
    
    return board, sol
```

```python
from collections import deque

def find_path(board, target):
    start = (0, 0)
    if start == target:
        return [start]
    
    n = max(max(x for x, y in board.keys()), max(y for x, y in board.keys())) + 1
    
    visited = set()
    queue = deque()
    queue.append((start, [start]))
    visited.add(start)
    
    while queue:
        current_pos, path = queue.popleft()
        x, y = current_pos
        
        direction = board.get(current_pos)
        if not direction:
            continue
        
        next_cells = []
        if direction == 'N':
            next_cells = [(x, new_y) for new_y in range(y-1, -1, -1)]
        elif direction == 'S':
            next_cells = [(x, new_y) for new_y in range(y+1, n)]
        elif direction == 'W':
            next_cells = [(new_x, y) for new_x in range(x-1, -1, -1)]
        elif direction == 'E':
            next_cells = [(new_x, y) for new_x in range(x+1, n)]
        
        for next_cell in next_cells:
            if next_cell == target:
                return path + [next_cell]
            if next_cell not in visited:
                visited.add(next_cell)
                queue.append((next_cell, path + [next_cell]))
    
    return None

# Test Case 1: Generate an 8x8 board solvable in 4 steps
print("Test Case 1: 8x8 Board with Solution in 4 Steps")
board1, sol1 = create_game(8, 4)
target1 = sol1[-1]
path1 = find_path(board1, target1)
print("Computed Path:", path1)
print("-" * 50)

# Test Case 2: Generate another 8x8 board solvable in 5 steps
print("Test Case 2: 8x8 Board with Solution in 5 Steps")
board2, sol2 = create_game(8, 5)
target2 = sol2[-1]
path2 = find_path(board2, target2)
print("Computed Path:", path2)
print("-" * 50)
```

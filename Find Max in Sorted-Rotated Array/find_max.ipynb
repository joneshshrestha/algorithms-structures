```python
from math import floor
from random import randrange
def rnd_L(n):
    'create random Lambda list of n distinct elements'
    lst = [randrange(1,floor(n**1.5)) for _ in range(n+1)]
    max_ind = randrange(n+1)
    for i in range(max_ind):
        lst[i] += lst[i-1]
    for i in range(max_ind, n+1):
        lst[i] = lst[i-1] - lst[i]
    lst = [lst[i] + abs(lst[n]) for i in range(n)]
    return lst
```

```python
import itertools
def find_max(lst):
    
    # Base case
    # Handling empty list
    if len(lst) == 0:  
        return 'The list is empty.'
    # Handling single element
    elif len(lst) == 1:  
        return lst[0]
    # Handling two elements
    elif len(lst) == 2:  
        if lst[0] > lst[1]:
            return lst[0]
        else:
            return lst[1]
            
    # Initializing the left and right pointers
    left = 0
    right = len(lst) - 1

    # Loop until right pointer meets or exceeds the left pointer, meaning the search operations is complete
    while left <= right:
        # Set the mid pointer
        mid = (left + right) // 2

        # Edge case handling for when the peak is at the start and when peak is at the end
        
        # When mid is start and peak is at the start, returns the peak
        # If not peak must be to the right so increase the left to mid + 1 
        if mid == 0:
            if lst[0] > lst[1]:
                return lst[0]
            else:
                left = mid + 1
            continue
        # When mid is at the end and peak is at the end, returns the peak
        # If not peak must be to the left so increase the right pointer to mid + 1
        if mid == len(lst) - 1:
            if lst[mid] > lst[mid-1]:
                return lst[mid]
            else: 
                right = mid - 1
            continue
            
        # Check if mid is the peak (max element) and return if it is the peak.
        if lst[mid] > lst[mid-1] and lst[mid] > lst[mid+1]:
            return lst[mid]

        # Check if mid is less than the right element, meaning right has the peak
        if lst[mid] < lst[mid+1]:
            left = mid + 1
        # Check if mid is less than the left element, meaning left has the peak
        elif lst[mid] < lst[mid-1]:
            right = mid - 1
```
